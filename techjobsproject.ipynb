```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df [var],kde= True) 


plotting('job_id',1)
plotting('company_rating',2)
plotting('experience_min_yrs',3)
plotting('experience_max_yrs',4)


plt.tight_layout()
```

```python
df['experience_min_yrs'].value_counts()
```

```python
df_mean = df.loc[df['experience_min_yrs']!=0, 'experience_min_yrs'].mean() 
df_mean
```

```python
df['experience_min_yrs'] = df['experience_min_yrs'].replace(0,df_mean)
df['experience_min_yrs'] = df['experience_min_yrs'].round(2)
```

```python
df.head()
```

```python
df_mean = df.loc[df['experience_max_yrs']!=0, 'experience_max_yrs'].mean() 
df['experience_max_yrs'] = df['experience_max_yrs'].replace(0,df_mean)
df['experience_max_yrs'] = df['experience_max_yrs'].round(2)
```

```python
df.head()
```

```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df [var],kde= True) 


plotting('job_id',1)
plotting('company_rating',2)
plotting('experience_min_yrs',3)
plotting('experience_max_yrs',4)


plt.tight_layout()
```
